```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
data_set = pd.read_csv(
    '/Users/zolboo/Desktop/bachelor/employementAnalysis/project/2022-04-08adScrape_cleaned.csv', error_bad_lines=False)
print(len(data_set))
data_set.head(10)
```

```python
sentences = []
y = []
z = []

for index, row in data_set.iterrows():
  title = row['jobTitle']
  employee = row['employee']
  sentences.append(str(title) + ' ' + str(employee))
  z.append(row['minSalary'])
  y.append(row['type'])
```

```python
sen_model = SentenceTransformer('sentence-transformers/msmarco-distilbert-multilingual-en-de-v2-tmp-lng-aligned')
```

```python
embeddings = sen_model.encode(sentences)
```

```python
x = embeddings
```

```python
it = 0 
x_ = []
for i in range(len(x)):
  B = np.array([z[it]])
  x_.append(np.append(x[i], B))
  it += 1
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x_, y)
```
